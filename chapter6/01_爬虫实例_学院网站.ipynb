# 1 学院网站学术活动通知爬虫（静态网页爬虫）

## 1.1 单个页面的获取

```python
#导入爬虫需要的包
import requests
from bs4 import BeautifulSoup as bs

#设置网站地址, 获取网站信息
url = 'http://sf.cufe.edu.cn/kydt/xshd.htm' 
res = requests.get(url)
soup = bs(res.content, 'lxml')
#print(soup)

#解析网页信息

#1.找出class属性值为list的ul(回想HTML语言的结构)
news_list = soup.find('ul', {'class': 'list'})

#2.找出ul标签下的所有li标签
news = news_list.find_all('li')
news_titles = []
news_source = []

#3.遍历news
for i in news:
    try:
        # 提取新闻标题
        title = i.find('a').get_text().strip()
        # 提取新闻来源
        source = 'http://sf.cufe.edu.cn/' + i.find('a')['href'][2:]
        # 存储爬取结果
        news_titles.append(title)
        news_source.append(source)
        print('新闻标题：', title)
        print('新闻来源：', source)
        print()
    except Exception as e:
        print('似乎出现了错误!')
        print(e)
        continue
```

## 1.2 循环获取所有的页面

```python
#导入爬虫需要的包
import requests
from bs4 import BeautifulSoup as bs

#设置网站地址, 获取网站信息
url = 'http://sf.cufe.edu.cn/kydt/xshd.htm' 
res = requests.get(url)
soup = bs(res.content, 'lxml')

#获取页码数量
page_number = soup.find('td', {'id': 'fanye234303'}).get_text().strip()[-2:]

page_number = int(page_number)

#循环获取所有页面的信息
for i in range(page_number - 1):
    if i == 0:
        
        url = 'http://sf.cufe.edu.cn/kydt/xshd.htm' 
        
    else:
        
        url = 'http://sf.cufe.edu.cn/kydt/xshd/' + str(page_number - i - 1) + '.htm' 
        
    res = requests.get(url)

    
    #解析网页信息
    #注意这里是res.content而不是res.text
    soup = bs(res.content, 'lxml')

    #1.找出class属性值为news-list的div
    news_list = soup.find('ul', {'class': 'list'})

    #2.找出news_list下的所有li标签
    news = news_list.find_all('li')
    news_titles = []
    news_source = []

    #3.遍历news
    for i in news:
        try:
            # 提取新闻标题
            title = i.find('a').get_text().strip()
            # 提取新闻来源
            source = 'http://sf.cufe.edu.cn/' + i.find('a')['href']
            # 存储爬取结果
            news_titles.append(title)
            news_source.append(source)
            print('新闻标题：', title)
            print('新闻来源：', source)
            print()
        except Exception as e:
            print('似乎出现了错误!')
            print(e)
            continue
```
