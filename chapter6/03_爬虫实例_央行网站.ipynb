## 1.1 单个新闻发布页面的获取

>**初始设定**

>两个代码块中的path = '/Users/fanzhihao/Downloads/chromedriver'均需要改为自己Chrome Driver的安装路径

>Chrome Driver的版本必须与自己的Chrome浏览器版本完全一致

>使用Selenium时请保证网速较快, 网速过低会导致程序运行缓慢、还可能会触发Error

```python
#导入爬虫需要的包
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.webdriver.chrome.service import Service
import time


#设置chrome driver基本属性, 获取央行主站的信息

#若未设置好路径, 需要换成自己chrome driver的安装路径
path = '/Users/fanzhihao/Downloads/chromedriver'
#s = Service(path) 
browser = webdriver.Chrome(service = s)

#若已经设置好的路径, 则可讲前面两句注释掉, 使用下面的语句
#windows 放在\Users\lenovo\AppData\Local\Google\Chrome\Application
#mac 放在/user/local/bin
#browser = webdriver.Chrome()

url = "http://www.pbc.gov.cn/rmyh/index.html"
browser.get(url)

#暂停3秒, 等带浏览器加载完毕
time.sleep(3)

#找到新闻发布板块的xpath, 获取并进行点击
browser.find_element(by = 'xpath', value = '//*[@id="logo"]/div[1]/table/tbody/tr/td[2]/table/tbody/tr[1]/td[1]/a').click()

res = browser.page_source

#解析爬取的网页

soup = bs(res, 'lxml')

content= soup.find_all('font', {'class': 'newslist_style'})

news_titles = []
news_source = []

for i in content:
    
    title = i.find('a')['title']
    source = 'http://www.pbc.gov.cn' + i.find('a')['href']
    news_titles.append(title)
    news_source.append(source)
    
    print('新闻标题：', title)
    print('新闻来源：', source)
    print()
```

## 1.2 循环获取所有新闻发布信息

```python
#导入爬虫需要的包
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.webdriver.chrome.service import Service
import time


#设置chrome driver基本属性, 获取央行主站的信息

#若未设置好路径, 需要换成自己chrome driver的安装路径
path = '/Users/fanzhihao/Downloads/chromedriver'
s = Service(path) 
browser = webdriver.Chrome(service = s)

#若已经设置好的路径, 则可讲前面两句注释掉, 使用下面的语句
#browser = webdriver.Chrome()

url = "http://www.pbc.gov.cn/rmyh/index.html"
browser.get(url)

#暂停3秒, 等带浏览器加载完毕
time.sleep(3)


#找到新闻发布板块的xpth, 获取并进行点击
browser.find_element(by = 'xpath', value = '//*[@id="logo"]/div[1]/table/tbody/tr/td[2]/table/tbody/tr[1]/td[1]/a').click()

res = browser.page_source

#解析爬取的第一页的信息

soup = bs(res, 'lxml')

content= soup.find_all('font', {'class': 'newslist_style'})

news_titles = []
news_source = []

for i in content:
    
    title = i.find('a')['title']
    source = 'http://www.pbc.gov.cn' + i.find('a')['href']
    news_titles.append(title)
    news_source.append(source)
    
    print('新闻标题：', title)
    print('新闻来源：', source)
    print()
    
    
#解析爬取剩余页面的信息(前20页)

for i in range(20):
    
    browser.find_element(by = 'xpath', value = '//*[@id="11040"]/div[2]/div[2]/table/tbody/tr/td[1]/a[3]').click()
    
    #暂停3秒, 等带浏览器加载完毕
    time.sleep(1)
    
    res = browser.page_source

    #解析爬取的第一页的信息

    soup = bs(res, 'lxml')

    content= soup.find_all('font', {'class': 'newslist_style'})

    news_titles = []
    news_source = []

    for i in content:

        title = i.find('a')['title']
        source = 'http://www.pbc.gov.cn' + i.find('a')['href']
        news_titles.append(title)
        news_source.append(source)

        print('新闻标题：', title)
        print('新闻来源：', source)
        print()
```
