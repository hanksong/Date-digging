```python
#导入爬虫需要的包
import requests
from bs4 import BeautifulSoup as bs

#设置网站地址, 获取网站信息
url = 'http://sf.cufe.edu.cn/xyxw.htm' 
res = requests.get(url)
soup = bs(res.content, 'lxml')

#获取页码数量
page_number = 62

news_titles = []
news_source = []

#循环获取所有页面的信息
for i in range(page_number - 1):
    if i == 0:
        
        url = 'http://sf.cufe.edu.cn/xyxw.htm'
        
    else:
        
        url =  'http://sf.cufe.edu.cn/xyxw/' + str(page_number - i - 1) + '.htm' 
        
    res = requests.get(url)

    
    #解析网页信息
    #注意这里是res.content而不是res.text
    soup = bs(res.content, 'lxml')

    #1.找出class属性值为news-list的div
    news_list = soup.find('ul', {'class': 'list'})

    #2.找出news_list下的所有li标签
    news = news_list.find_all('li')


    #3.遍历news
    for i in news:
        try:
            # 提取新闻标题
            title = i.find('a').get_text().strip()
            # 提取新闻来源
            source = 'http://sf.cufe.edu.cn/' + i.find('a')['href']
            # 存储爬取结果
            news_titles.append(title)
            news_source.append(source)
            print('新闻标题：', title)
            print('新闻来源：', source)
            print()
        except Exception as e:
            print('似乎出现了错误!')
            print(e)
            continue
```

```python
f = open("学院新闻正文.doc","w",encoding = "utf-8")
i = 0
for newsurl in news_source:
    i = 1
    rget = requests.get(newsurl)
    soup = bs(rget.content,'lxml')
    f.write(str(i)+"、")
    f.write("标题："+ soup.h3.string + '\n')
    f.write("正文：")
    for tag in soup.find_all('p',style = 'text-indent: 2em;'):
        try:
            f.write(tag.string)
        except TypeError as Te:
            f.write("类型错误")
            continue
        except AttributeError  as Ae:
            f.write("页面丢失")
            continue
    f.write("\n")
f.close()
```
